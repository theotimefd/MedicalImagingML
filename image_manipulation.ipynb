```python
import pydicom
import matplotlib.pyplot as plt
```

```python
# Reading a DICOM file from a specific path
dcm_data = pydicom.dcmread('data/000000.dcm')
```

```python
print(dcm_data)
```

```python
modality = dcm_data.Modality
print(modality)

image_type = dcm_data.ImageType
print(image_type)
```

```python
image = dcm_data.pixel_array

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Axial Slice of a Chest-CT')
plt.show()
```
